```python
# Remove Columns: Transition, Transition.2 Education, Later, SSPG Date, Followup, Study 
merged_df = merged_df.drop(columns=['Transition', 'Transition.2', 'Education', 'Later', 'SSPG.Date', 'Followup', 'Study'])
display(merged_df)
# Save the results to a file
merged_df.to_csv('/Users/naifganadily/Desktop/Artificial Intelligence for Precision Nutrition Bootcamp/gut-microbiome-precision-nutrition/data/results/merged.hmp.h1.truth.cleaned.txt', sep='\t')
```

```python
# One Hot Encoding: Sex, Class, Race, IR_IS (Ground Truth)
from sklearn.preprocessing import OneHotEncoder
```

```python
# Add the columns to a list to be easily integrated
encoding_columns = ['Sex', 'Class', 'Race', 'IR_IS']
func_encoder = OneHotEncoder(sparse=False, drop='first')
encoded_columns = func_encoder.fit_transform(merged_df[encoding_columns])
encoded_columns = pd.DataFrame(encoded_columns, columns=func_encoder.get_feature_names_out(encoding_columns))

merged_df = merged_df.drop(columns=encoding_columns)
df_onehot = pd.concat([merged_df.reset_index(drop=True), encoded_columns.reset_index(drop=True)], axis=1)

display(df_onehot)
```

- 0 is Insulin Resistant
- 1 is Insulin Sensitive

- 0 is Female
- 1 is Male

```python
# StandardScaler: Age, BMI, SSPG
from sklearn.preprocessing import StandardScaler
```

```python
# This portion is to scale the columns Age, BMI, and SSPG using StandardScaler()
con_features = ['Age', 'BMI', 'SSPG']
scaler = StandardScaler()
scaled_values = scaler.fit_transform(df_onehot[con_features])
scaled_df = pd.DataFrame(scaled_values, columns=con_features)

# This portion is to drop the columns that were scaled then concatenate the scaled columns with the one hot encoded columns
df_onehot = df_onehot.drop(columns=con_features)
df_scaled_onehot = pd.concat([df_onehot.reset_index(drop=True), scaled_df.reset_index(drop=True)], axis=1)

display(df_scaled_onehot)
```
